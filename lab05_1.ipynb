```python
import random

#순차 탐색 함수 - 리스트의 처음부터 끝까 차례로 탐색, 
#찾으면 해당 인덱스를 반환하고 찾지 못하면 -1 반환 
def sequential_search(arr, target):
    targetIndex = 0
    for i in range(len(arr)):
        #현재 원소 찾는 값과 같다면 
        if arr[i] == target:
            #인덱스 저장 
            tagetIndex = i
            #반환, 탐색 종료 
            return tagetIndex
    #반복 종료 시까 찾지 못하면 -1 반환 
    return -1


n = int(input("생성할 랜덤 변수 개수를 입력하세요: "))

arr = random.sample(range(1, 100), n)
print("리스트:", arr)

target = int(input("target data를 입력하세요: "))

targetIdx = sequential_search(arr, target)

if targetIdx == -1:
    print(f"값 {target}을 찾을 수 없습니다.")
else:
    print(f"값 {target}은 인덱스 {targetIdx}에 있습니다.")
```
