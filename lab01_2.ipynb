```python
#버블정렬 
import random

def bubble_sort(lst, n):
    for tmp in range(1, n+1):
        #앞뒤 원소 비교 
        for i in range(1, n - tmp + 1):
            #앞 원소 뒤 원소보 크면
            if lst[i - 1] > lst[i]:
                #원소 교환 
                lst[i], lst[i - 1] = lst[i - 1], lst[i]
        print("단계 ", tmp, ": ", lst)

n = int(input("생성할 랜덤 변수 개수를 입력하세요: "))
#1~99 범위 정수 중 n개 중복 허용하여 랜덤 생성
num_list = random.choices(range(1, 100), k=n)

bubble_sort(num_list, len(num_list))

print("최종 정렬 결과: ", num_list)
```
