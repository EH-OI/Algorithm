```python
#퀵 정렬 
import random

#분할 함수 
def partition(lst, left, right):
    #첫 번째 원소를 pivot으로 선택 
    pivot_value = lst[left]
    pivot = left
    print("생성된 피벗의 값:", pivot_value)
    left += 1
    while True:
        #pivot보다 작은 값을 찾을 때까지 오른쪽으로 이동
        while left <= right and lst[left] < pivot_value:
            left += 1
        #pivot보다 큰 값을 찾을 때까지 왼쪽으로 이동 
        while left <= right and lst[right] > pivot_value:
            right -= 1
        #위치가 교차하지 않았다면 -> 두 값 교환 
        if left < right:
            lst[left], lst[right] = lst[right], lst[left]
            left += 1
            right -= 1
        else:
            break
    #pivot 올바른 위치 이동 
    lst[pivot], lst[right] = lst[right], lst[pivot]
    print("정렬 단계:",lst)
    #pivot 최종 위치 
    return right

def quick_sort(lst, left, right):
    if left < right:
        split = partition(lst, left, right) #분할 과정
        quick_sort(lst, left, split - 1) #왼쪽 부분 리스트 재귀 정렬 
        quick_sort(lst, split + 1, right) #오른쪽 부분 리스트 재귀 정렬 
    return lst

#리스트 크기 n 입력
n = int(input("생성할 랜덤 변수 개수를 입력하세요: "))
#1~99 범위 내 랜덤 수 n개 생성 
num_list = random.choices(range(1, 100), k=n)

print("생성된 list:", num_list)
print("퀵 정렬 시작")
sorted_list = quick_sort(num_list, 0, len(num_list) - 1)

print("최종 정렬 결과:", sorted_list)
```
