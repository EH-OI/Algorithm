```python
import random

n = int(input("생성할 랜덤 변수 개수를 입력하세요: "))

#중복 허용 난수 생성 
num_list = random.choices(range(1, 1001), k=n)
print("data List:", num_list)

#첫번째 값 = 비교기준값 
max_num = num_list[0]
min_num = num_list[0]

#선형탐색
for i in num_list:
    #현재 값이 최댓값보다 클 때 
    if i > max_num:
        max_num = i
    #현재 값이 기존 최솟값보다 작을 때 
    if i < min_num:
        min_num = i

print(f"최대값: {max_num}, 최솟값: {min_num}")
```
