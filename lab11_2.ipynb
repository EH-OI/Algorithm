```python
def dfs(v, graph, visited):
    #현재 노드를 방문 처리하고 출력
    visited[v] = True
    print(f"현재 방문 중인 노드의 번호 : {v}")
    
    #현재 노드와 연결된 다른 노드를 재귀적 방문
    for i in graph[v]:
        if not visited[i]:
            dfs(i, graph, visited)

#1. 노드 개수 입력
n = int(input("노드의 개수를 입력하세요 : "))

#그래프 초기화 
graph = [[] for _ in range(n)]

#2. 각 노드의 인접 노드 정보 입력
for i in range(n):
    print("인접한 노드 입력을 종료하시려면 -1을 입력하세요")
    while True:
        try:
            node = int(input(f"{i}노드에 인접한 노드를 입력하세요 : "))
            
            if node == -1:
                break
            #인접 노드 추가
            graph[i].append(node)
        except ValueError:
            pass #예외처리

#3. 그래프 정보 출력
print("\n***********그래프 정보 출력!**********")
for i in range(n):
    print(f"{i}번째 노드의 인접한 노드 :", *graph[i])

#4. DFS 탐색
print("\n***********DFS 시작!**********")
start_node = int(input("탐색시작 지점을 입력하세요 : "))
print("탐색 순서는 다음과 같습니다.")

# 방문 여부를 저장 리스트 
visited = [False] * n

dfs(start_node, graph, visited)
```
