```python
import random

def multiway_merge_sort(list, group_num):
    #리스트 크기가 1 이하라면 
    if len(list) <= 1:
        #이미 정렬된 상태 
        return list

    #리스트를 group_num개 그룹으로 순차 분할
    groups = [[] for _ in range(group_num)]
    for i in range(len(list)):
        groups[i % group_num].append(list[i])

    print("분할된 리스트 그룹:", groups)

    #분할된 각 그룹에 multiway_merge_sort 재귀적 적용 -> 부분 정렬 
    sorted_groups = []
    for group in groups:
        sorted_group = multiway_merge_sort(group, group_num)
        sorted_groups.append(sorted_group)

    #정렬된 그룹 병합
    return merge_sorted_groups(sorted_groups)

#단순 삽입
def insertion_sort(list):
    for i in range(1, len(list)):
        current = list[i]
        j = i - 1
        while j >= 0 and list[j] > current:
            list[j + 1] = list[j]
            j -= 1
        list[j + 1] = current
    return list

#병합 단계 
def merge_sorted_groups(groups):
    print("정렬 후 병합 대상 그룹:", groups)
    merged = sum(groups, []) #2차원 리스트를 1차원 리스트로 병합 
    return insertion_sort(merged) #완전 정렬 


# 메인 실행
n = int(input("생성할 랜덤 변수 개수를 입력하세요: "))
group_num = int(input("몇 개의 그룹으로 나눌 것인지 입력하세요: "))
num_list = random.sample(range(1, 1000), k=n) #중복을 허용하지 않는 난수 
print("랜덤 데이터 리스트:", num_list)

final_sorted = multiway_merge_sort(num_list, group_num)
print("최종 정렬된 리스트:", final_sorted)
```
