```python
def power_of_two(n, depth=1):
    #현재 분할 단계 출력
    print(f"{depth}차 분할: 2^{n}")

    #기본 조건: n=0이라면 2^0=1 반환 
    if n == 0:
        print(f"{depth}차 합치기: 2^0 = 1")
        return 1

    #n=1이라면 추가 분할 없이 결과 반환 
    if n == 1:
        print(f"{depth}차 분할: 2^1")
        print(f"{depth}차 합치기: 2^1 = 2^0 * 2^0 * 2 = 2")
        return 2

    #분할정복: 지수를 절반으로 나누어 재귀 호출 
    half = power_of_two(n // 2, depth + 1)

    #n이 짝수라면 2^n = (2^(n/2))^2
    if n % 2 == 0:
        result = half * half
        print(f"{depth}차 합치기: 2^{n} = 2^{n//2} * 2^{n//2} = {result}")
    else:
        #n이 홀수라면 2^n=(2^(n/2))^2*2
        result = half * half * 2
        print(f"{depth}차 합치기: 2^{n} = 2^{n//2} * 2^{n//2} * 2 = {result}")

    return result


# 실행부
N = int(input("N의 값을 입력해주세요: "))
print("분할정복 시작")
result = power_of_two(N)
print(f"\n2의 {N}제곱은 {result}입니다.")
```
