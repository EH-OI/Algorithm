```python
#합병 정렬 
import random

#재귀적 합병 정렬 
def merge_sort(lst):
    #리스트 크기가 1 이하이면 -> 이미 정렬된 상태 
    if len(lst) <= 1:
        return lst

    #리스트 절반 분할 
    mid = len(lst) // 2
    left = lst[:mid]
    right = lst[mid:]

    #각각 재귀적으로 정렬 
    left_divided = merge_sort(left)
    right_divided = merge_sort(right)
    #분할 상태 출력 
    print(lst)
    #두 부분 배열 병합 
    return merge(left_divided, right_divided)

#반복적 합병 정렬 
def iterative_merge_sort(lst):
    n = len(lst)
    #1개 단위로 병합 시작 
    width = 1

    #width가 리스트 크기보다 작을 동안 
    while width < n:
        #현 width 단위로 리스트 순차적 병합 
        for i in range(0, n, 2 * width):
            left = lst[i:i + width] #왼쪽 부분 배열 
            right = lst[i + width:i + 2 * width] #오른쪽 부분 배열 
            #두 배열 병합 후 리스트에 재삽입 
            lst[i:i + 2 * width] = merge(left, right)
            
        #각 단계별 정렬 상태 출력
        print("크기", width, "단계 :", lst)
        #width 두 배 증가 
        width *= 2

    return lst

#두 정렬된 배열 병합 
def merge(left, right):
    result = []
    i, j = 0, 0
    #두 배열 중 작은 값을 결과 리스트에 추 
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    #남은 원소들을 결과 리스트에 추가 
    result.extend(left[i:])
    result.extend(right[j:])

    return result
    
#리스트 크기 n 입력
n = int(input("생성할 랜덤 변수 개수를 입력하세요: "))
#1~99 범위 내 랜덤 수 n개 생성 
num_list = random.sample(range(1, 100), n)
print("생성된 List:", num_list)

sorted_list = iterative_merge_sort(num_list)
print("정렬 완료된 List:", sorted_list)
```
