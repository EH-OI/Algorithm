```python
#선택정렬 
import random

def selection_sort(lst, n):
    for i in range(n):
        min_index = i #가장 작은 원소 인덱 저장 

        #i 이후 원소 중 최솟값 찾기 
        for j in range(i + 1, n):
            if lst[j] < lst[min_index]:
                min_index = j
        #최솟값이 현 위치와 다를 때
        if min_index != i:
            lst[i], lst[min_index] = lst[min_index], lst[i]  #i와 교환 
        print("단계 ", i+1, ": ", lst)

n = int(input("생성할 랜덤 변수 개수를 입력하세요: "))

#1~99 범위 정수 중 n개 중복 허용하여 랜덤 생성
num_list = random.choices(range(1, 100), k=n)

selection_sort(num_list, len(num_list))
print("최종 정렬 결과 : ", num_list)
```
