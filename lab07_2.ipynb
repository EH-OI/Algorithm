```python
def run_boyer_moore_search(pat, txt):

    #문자별 마지막 등장 위 테이블 생성 
    def build_bad_char_table(pat):
        table = [-1] * 256
        for idx in range(len(pat)):
            table[ord(pat[idx])] = idx
        return table

    #Boyer-Moore 탐색
    #Bad Character 규칙 
    def search_with_boyer_moore(pat, txt):
        pat_len = len(pat)
        txt_len = len(txt)
        bad_char_table = build_bad_char_table(pat)
        shift = 0 #텍스트에서 패턴 시작 위치
        while shift <= txt_len - pat_len:
            compare_idx = pat_len - 1 #패턴의 끝부터 비교 
            #오른쪽부터 일치하는 동안 왼쪽으로 이동 
            while compare_idx >= 0 and pat[compare_idx] == txt[shift + compare_idx]:
                compare_idx -= 1 

            #패턴 전체 일치 시
            if compare_idx < 0:
                print(f"패턴이 인덱스 {shift} 에서 발견되었습니다.")
                print("해당 패턴을 찾았습니다.\n")
                return shift
            else:
                #불일치 시 이동 거리 계산 
                shift += max(1, compare_idx - bad_char_table[ord(txt[shift + compare_idx])])
        #탐색 실패 
        print("해당 패턴을 찾지 못했습니다.\n")
        return -1

    return search_with_boyer_moore(pat, txt)


txt_data = "Human Compter Interection Lab, Jin-Woo Jung, Algorithm Lab Work6"
print("예시 문자열 :", txt_data)

for trial in range(3):
    print("입출력 예제", trial + 1)
    user_pattern = input("찾을 패턴을 입력하세요 : ")
    result = run_boyer_moore_search(user_pattern, txt_data)
```
