```python
# Rabin-Karp Algorithm

def rabin_karp(arr, target):
    targetLen = len(target)
    arrLen = len(arr)

    #패턴이 전체 문자열보다 길면 검색 불가능 
    if targetLen > arrLen:
        return -1

    base = 256 
    prime = 101
    targetHash = 0
    arrHash = 0
    h = 1

    #h = (base^(targetLen-1)) % prime 계산
    for i in range(targetLen - 1):
        h = (h * base) % prime

    #초기 해시 계산 
    for i in range(targetLen):
        targetHash = (base * targetHash + ord(target[i])) % prime
        arrHash = (base * arrHash + ord(arr[i])) % prime

    #윈도우를 한 칸씩 이동하며 비교 
    for i in range(arrLen - targetLen + 1):
        #해시값이 같으면 실제 문자열 비교
        if targetHash == arrHash:
            if arr[i:i + targetLen] == target: #실제 비교
                return i

        #다음 윈도우 해시 갱신
        if i < arrLen - targetLen:
            arrHash = (base * (arrHash - ord(arr[i]) * h) + ord(arr[i + targetLen])) % prime

    return -1 #패턴이 없는 경우 

str_ex = "Human Compter Interection Lab, Jin-Woo Jung, Algorithm Lab Work6"

str_find = input("찾을 패턴을 입력하세요: ")

find_idx = rabin_karp(str_ex, str_find)
if find_idx == -1:
    print("해당 패턴을 찾지 못했습니다.")
else:
    print(f"패턴이 인덱스 {find_idx}에서 발견되었습니다.")
    print("해당 패턴을 찾았습니다.")
```
