```python
import random
import math

def insertion_sort(arr):
    #버킷 내부 원소를 정렬 -> 삽입 정렬 사용 
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        #key보다 큰 값들을 오른쪽으로 이동 
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        #key를 올바른 위치에 삽입 
        arr[j + 1] = key
    return arr

def bucket_sort(lst):
    if not lst:
        return []

    max_n = max(lst) #리스트 최댓값 
    min_n = min(lst) #리스트 최솟값 
    bucket_count = len(lst) #버킷 개수 = 입력 크기

    #모든 값이 같은 경우
    if max_n == min_n:
        return lst #이미 정렬된 상태 

    #각 버킷 크기 = (최댓값 - 최솟값) / (버킷 개수 - 1)
    bucket_size = (max_n - min_n) / (bucket_count - 1)
    buckets = [[] for _ in range(bucket_count)] #빈 버킷 리스트

    #원소들을 버킷에 분배 
    for num in lst:
        bucket_index = math.floor((num - min_n) / bucket_size)
        #최댓값이 마지막 버킷에 들어가도록 
        if bucket_index == bucket_count:
            bucket_index = bucket_count - 1
        buckets[bucket_index].append(num)
        #각 원소가 분배된 이후의 버킷 상태 
        print(buckets)

    #각 버킷 정렬, 합치기 
    sorted_list = []
    for bucket in buckets:
        #각 버킷에 삽입정렬 적용 
        sorted_list.extend(insertion_sort(bucket))

    return sorted_list

n = int(input("생성할 랜덤 변수 개수를 입력하세요: "))

num_list = random.choices(range(1, 101), k=n)
print("생성된 list:", num_list)

sorted_list = bucket_sort(num_list)
print("최종 정렬 결과:", sorted_list)
```
