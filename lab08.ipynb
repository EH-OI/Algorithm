```python
def getLineInfo(x1, y1, x2, y2):
    #두 점으로 직선 정보 계산 
    if x1 == x2:
        if y1 == y2:
            return None #점이 1개 -> 직선 아님
        return "vertical", x1 #수직선 x = 상수
    else:
        slope = (y2 - y1) / (x2 - x1) #기울기
        intercept = y1 - slope * x1 #y절편 
        return slope, intercept


def getLineRelation(l1, l2):
    #두 직선 관계 판단 
    if l1 is None or l2 is None:
        return "에러: 직선이 생기지 않습니다."

    #둘 다 수직선 
    if l1 == "vertical" and l2 == "vertical":
        if l1(x1) == l2(x1):
            return "두 직선은 일치합니다."
        else:
            return "두 직선은 평행합니다."

    #한쪽만 수직선 
    elif l1 == "vertical" or l2 == "vertical":
        return "두 직선은 교차합니다."

    #일반 직선 비교 
    slp1, ic1 = l1
    slp2, ic2 = l2

    if slp1 == slp2:
        if ic1 == ic2:
            return "두 직선은 일치합니다."
        else:
            return "두 직선은 평행합니다."
    else:
        return "두 직선은 교차합니다."

#첫 번째 직선 입력
print("첫번째 직선")
x1, y1 = map(float, input("첫번째 포인트의 x,y를 입력하시오 : ").split())
x2, y2 = map(float, input("두번째 포인트의 x,y를 입력하시오 : ").split())
line1 = getLineInfo(x1, y1, x2, y2)

#두 번째 직선 입력
print("두번째 직선")
x3, y3 = map(float, input("첫번째 포인트의 x,y를 입력하시오 : ").split())
x4, y4 = map(float, input("두번째 포인트의 x,y를 입력하시오 : ").split())
line2 = getLineInfo(x3, y3, x4, y4)

#결과
relation = getLineRelation(line1, line2)
print(relation)
```
