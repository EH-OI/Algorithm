```python
import heapq

def dijkstra(graph, source, target):
    #거리 및 이전 노드 초기화
    cost = {node: float('inf') for node in graph}
    cost[source] = 0
    route = {node: None for node in graph}
    queue = [(0, source)]

    while queue:
        curr_cost, curr_node = heapq.heappop(queue)

        #더 짧은 경로가 이미 있는 경우 
        if curr_cost > cost[curr_node]:
            continue #스킵 

        for adj, weight in graph[curr_node].items():
            new_cost = curr_cost + weight
            if new_cost < cost[adj]:
                cost[adj] = new_cost
                route[adj] = curr_node
                heapq.heappush(queue, (new_cost, adj))

    #경로 재구성
    trace = []
    node = target
    while node:
        trace.insert(0, node)
        node = route[node]

    return cost[target], trace


graph = {
    'A': {'B': 1, 'C': 5},
    'B': {'A': 1, 'C': 2, 'D': 6},
    'C': {'A': 5, 'B': 2, 'D': 1},
    'D': {'B': 6, 'C': 1}
}

#입력
start = input("시작 노드를 입력하세요: ").strip()
end = input("도착 노드를 입력하세요: ").strip()

#최단 경로 계산
min_cost, path_taken = dijkstra(graph, start, end)

print(f"출발 노드 {start}로부터 도착 노드 {end}까지의 최단 거리: {min_cost}")
print(f"최단 경로: {' -> '.join(path_taken)}")
```
