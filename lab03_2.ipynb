```python
import random

def counting_sort(arr, max_val):
    #등장 횟수를 저장할 count 배열 생성
    count_arr = [0] * (max_val + 1) 

    #주어진 배열의 각 원소에 대해 등장 횟수 기록 
    for num in arr:
        count_arr[num] += 1
    print("초기 원소의 개수 배열 :", count_arr[1:]) #인덱스 1부터 출력 
    temp_count = count_arr[:] #이후 과정 출력을 위한 복사본 생성 

    #count 배열을 바탕으로 정렬된 결과 생성 
    idx = 1 #현재 채울 값 

    #원본 배열의 길이만큼 반복 
    for i in range(len(arr)):
        #현재 값의 개수 = 0 -> 다음 값 이동 
        while temp_count[idx] == 0:
            idx += 1

        #정렬된 순서대로 원소 배치 
        arr[i] = idx
        temp_count[idx] -= 1
        
        print("원소의 개수 배열 :", temp_count[1:])
        print("단계별 배열 :", arr)

    return arr


num_elements = int(input("생성할 랜덤 변수 개수를 입력하세요: "))
random_list = random.choices(range(1, 11), k=num_elements)
#리스트 최댓값을 정렬 범위로 설정
max_value = max(random_list)

print("생성된 List:", random_list)
sorted_list = counting_sort(random_list, max_value)
print("정렬 완료된 List:", sorted_list)
```
