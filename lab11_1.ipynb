```python
def edit_distance(n, X, m, Y, ins, delete, change):
    #테이블 D n+1, m+1 크기로 초기화 
    #D[i][j]=X의 처음 i글자를 Y의 처음 j글자로 바꾸는 최소 비용 
    D = [[0] * (m + 1) for _ in range(n + 1)]

    #1열 초기화 
    for i in range(1, n + 1):
        D[i][0] = D[i - 1][0] + delete

    #1행 초기화 
    for j in range(1, m + 1):
        D[0][j] = D[0][j - 1] + ins

    #DP 테이블 채우기 
    for i in range(1, n + 1):
        for j in range(1, m + 1):

            #문자가 같으면 교체 비용=0, 다르면 change 비용 적용
            c = 0 if X[i - 1] == Y[j - 1] else change

            #세 연산 중 최소비용 선택
            D[i][j] = min(D[i - 1][j] + delete, #X의 i번째 문자 삭제
                          D[i][j - 1] + ins, #Y의 j번째 문자 삽입
                          D[i - 1][j - 1] + c) #문자 교체 또는 그대로 사용 
    return D

x_input = input("초기 문자열을 입력하시오 : ")
y_input = input("목표 문자열을 입력하시오 : ")

n = len(x_input)
m = len(y_input)

#삽입, 삭제, 교체 비용 =1
table = edit_distance(n, x_input, m, y_input, 1, 1, 1)

print("\n[출력 예제]")
print("편집거리 테이블")

for row in table:
    print("  ".join(map(str, row)))

print(f"최소편집비용 : {table[n][m]}")
```
