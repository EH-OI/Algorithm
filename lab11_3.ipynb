```python
from collections import deque

def bfs(start_node, graph, n):
    #1. 방문 여부를 저장할 리스트 False로 초기화
    visited = [False] * n
    
    #2. 큐 생성 및 시작 노드 삽입
    queue = deque([start_node])
    
    #시작 노드 방문 처리
    visited[start_node] = True
    
    print("\n***********BFS 시작!**********")
    print("탐색 순서는 다음과 같습니다.")
    
    #3. 큐가 빌 때까지 반복
    while queue:
        #큐에서 노드 하나 꺼냄
        v = queue.popleft()
        print(f"현재 방문 중인 노드의 번호: {v}")
        
        #꺼낸 노드와 인접한 노드 확인
        for i in graph[v]:
            #아직 방문하지 않은 인접 노드라면
            if not visited[i]:
                #큐에 삽입, 방문 처리
                queue.append(i)
                visited[i] = True

#1. 노드 개수 입력
n_input = input("노드의 개수를 입력하세요: ")
n = int(n_input)

#그래프 초기화
graph = [[] for _ in range(n)]

#2. 각 노드의 인접 노드 정보 입력
for i in range(n):
    print("인접한 노드 입력을 종료하시려면 -1을 입력하세요")
    while True:
        try:
            line = input(f"{i} 노드에 인접한 노드를 입력하세요: ")
            node = int(line)
            
            if node == -1:
                break
            
            graph[i].append(node)
        except ValueError:
            pass

#3. 그래프 정보 출력
print("\n***********그래프 정보 출력!**********")
for i in range(n):
    print(f"{i}번째 노드의 인접한 노드:", *graph[i])

#4. 탐색 시작 지점 입력, BFS 실행
start_node_input = input("탐색 시작 지점을 입력하세요: ")
start_node = int(start_node_input)

bfs(start_node, graph, n)
```
