```python
#삽입정렬 
import random

def insertion_sort(lst, n):
    for i in range(1, n):
        value = lst[i] #현재 삽입할 원소 
        j = i #삽입할 위치 탐색 및 인덱스 이동 
        print("단계별 선택된 key :", value)
        #현재 삽입할 원소보다 큰 원소들을 오른쪽으로 이동 
        while j > 0 and lst[j-1] > value:
            lst[j] = lst[j-1]
            j -= 1
        #원소 삽입 
        lst[j] = value
        print(f"단계 {i}:", lst)

#리스트 크기 n 입력
n = int(input("생성할 랜덤 변수 개수를 입력하세요: "))
#1~99 범위 내 랜덤 수 n개 생성 
num_list = random.choices(range(1, 100), k=n)

print ("생성된 list :", num_list)
print("선택 정렬 시작")
insertion_sort(num_list, len(num_list))

print("최종 정렬 결과:", num_list)
```
