```python
import random

#이진탐색 함수 - 정렬된 리스트를 전제, 중간값을 기준으 탐색 범위를 절반씩 줄이 target 탐색 
def binary_search(arr, targetData, start, end):
    #현재 탐색 중인 구간 출력
    print("찾는 구간")
    #탐색 범위 내 리스트 출력 
    print(arr[start:end])

    #탐색 범위가 유효하 않은 경우
    if start >= end:
        return -1 #탐색 실패

    #중간 인덱스 계산 
    mid = (start + end) // 2 

    #중간값 = 목표값 
    if arr[mid] == targetData:
        #인덱스 반환 
        return mid

    #목표값 < 중간값 
    elif arr[mid] > targetData:
        #왼쪽 구간 재귀적 탐색 
        return binary_search(arr, targetData, start, mid)

    #목표값 > 중간값 
    else:
        #오른쪽 구간 재귀적 탐색 
        return binary_search(arr, targetData, mid + 1, end)


n = int(input("생성할 랜덤 변수 개수를 입력하세요: "))

arr = sorted(random.sample(range(1, 100), n))
print("리스트:", arr)

target = int(input("target data를 입력하세요: "))

targetIdx = binary_search(arr, target, 0, len(arr))

if targetIdx == -1:
    print(f"Target {target}을 찾을 수 없습니다.")
else :
    print(f"Target {target}은 인덱스 {targetIdx}에 있습니다.")
```
