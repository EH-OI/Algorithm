```python
#셸 정렬 
import random

def shell_sort(lst, n):
    #초기 간격 = 리스트 길이의 1/2
    gap = n // 2
    while gap > 0:
        #간격만큼 떨어진 원소들을 삽입정렬
        for i in range(gap, n):
            value = lst[i]
            j = i
            #gap 간격으로 떨어진 원소 비교, 이동 
            while j >= gap and lst[j-gap] > value:
                lst[j] = lst[j-gap]
                j -= gap
            #적절한 위치에 삽입
            lst[j] = value

        #각 단계별 정렬 상태 출력
        print(f"간격 {gap} 단계:", lst)
        #간격을 절반으로 줄임 
        gap //= 2

#리스트 크기 n 입력
n = int(input("생성할 랜덤 변수 개수를 입력하세요: "))
#1~99 범위 내 랜덤 수 n개 생성 
num_list = random.choices(range(1, 100), k=n)

print ("생성된 list :", num_list)
print("선택 정렬 시작")
shell_sort(num_list, len(num_list))

print("최종 정렬 결과:", num_list)
```
