```python
def search_pattern_kmp(pat, txt):
    def generate_lps_array(pat):
        len_lps = 0 #이전까지 최장 접두 = 접미사 길이 
        lps_arr = [0] * len(pat) #LPS 배열 초기화 
        idx = 1 #패턴 두 번째 문자부터 검사 시작 

        while idx < len(pat):
            if pat[idx] == pat[len_lps]:
                #접두사 = 접미사 
                len_lps += 1
                lps_arr[idx] = len_lps
                idx += 1
            else:
                #불일치 -> 이전 접두사 길이 
                if len_lps != 0:
                    len_lps = lps_arr[len_lps - 1]
                else:
                    #더이상 일치하 않을 경우 -> 0으로 초기화 
                    lps_arr[idx] = 0
                    idx += 1
        return lps_arr

    def execute_kmp(pat, txt):
        m = len(pat)
        n = len(txt)
        lps = generate_lps_array(pat) #LPS 배열 계산 
        idx_txt = 0 #텍스트 인덱스 
        idx_pat = 0 #패턴 인덱스 

        while idx_txt < n:
            #문자 일치 -> 양쪽 인덱스 증가 
            if pat[idx_pat] == txt[idx_txt]:
                idx_txt += 1
                idx_pat += 1

            #패턴 전체가 일치한 경우 
            if idx_pat == m:
                print(f"패턴이 인덱스 {idx_txt - idx_pat} 에서 발견되었습니다.")
                print("해당 패턴을 찾았습니다.\n")
                return idx_txt - idx_pat

            #불일치 시 -> LPS 배열 사용해 점프 
            elif idx_txt < n and pat[idx_pat] != txt[idx_txt]:
                if idx_pat != 0:
                    idx_pat = lps[idx_pat - 1]
                else:
                    idx_txt += 1

        print("해당 패턴을 찾지 못했습니다.\n")
        return -1

    return execute_kmp(pat, txt)


text_data = "Human Compter Interection Lab, Jin-Woo Jung, Algorithm Lab Work6"
print("예시 문자열 :", text_data)

for example in range(3):
    print("입출력 예제", example + 1)
    pattern_input = input("찾을 패턴을 입력하세요 : ")
    result = search_pattern_kmp(pattern_input, text_data)
```
