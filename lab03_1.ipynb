```python
import random

def heapsort(arr):
    length = len(arr)
    #주어진 배열을 힙 구조 변환 
    for idx in range(length // 2 - 1, -1, -1):
        #배열의 중간부터 역순으로 heapify 호출 
        heapify(arr, idx, length)

    #힙 정렬 - 가장 큰 값을 맨 뒤로 보내 힙 크 줄여가면서 반복 
    for idx in range(length - 1, 0, -1):
        #현재 힙의 최대값과 마지막 원소 교환 
        arr[idx], arr[0] = arr[0], arr[idx]
        #힙 크기 줄여 다시 최대 힙 유지
        heapify(arr, 0, idx)
        
        print("heap root:", arr[0]) #교환 뒤 새로운 힙 루트 
        print("정렬 단계:", arr) #현재 배열 상태 
    return arr

def heapify(arr, root, size):
    max_idx = root
    left_child = 2 * root + 1 #왼쪽 자식 인덱스 
    right_child = 2 * root + 2 #오른쪽 자식 인덱스 

    #왼쪽 자식이 존재하고, 루트보다 값이 클 때
    if left_child < size and arr[left_child] > arr[max_idx]:
        #max_idx 갱신 
        max_idx = left_child

    #오른쪽 자식이 존재하고, 현재까지의 최대값보다 클 때 
    if right_child < size and arr[right_child] > arr[max_idx]:
        #max_idx 갱신 
        max_idx = right_child

    #자식 노드 중 더 큰 값이 있을 때 swap 후 재귀적으 heapify 
    if max_idx != root:
        arr[root], arr[max_idx] = arr[max_idx], arr[root]
        #바뀐 자식 노드에서 다시 검사 
        heapify(arr, max_idx, size)

#실행 
num = int(input("생성할 랜덤 변수 개수를 입력하세요: "))
rand_list = random.sample(range(1, 101), k=num) #중복 없는 랜덤 숫자 생성
print("생성된 리스트:", rand_list)

sorted_list = heapsort(rand_list)
print("최종 정렬 결과:", sorted_list)
```
